# 1. What is Pandas? Why use it?

Pandas is a Python library used for data manipulation and analysis.

It provides high-performance, easy-to-use data structures (Series and DataFrame).

Widely used in data analysis, data cleaning, exploration, and machine learning preprocessing.

# 2. Why Pandas is powerful:

Handles large datasets efficiently.

Provides tools to filter, group, merge, and analyze data easily.

Works well with other libraries like NumPy, Matplotlib, Seaborn, and scikit-learn.

# 3. Series in Pandas

A 1-dimensional labeled array.

Can hold integers, floats, strings, or other Python objects.

Each element has an index.

```python
import pandas as pd

# Create a Series from a list
data = [10, 20, 30, 40]
series = pd.Series(data)
series

# Series with custom index:
series_1 = pd.Series(data, index=['a', 'b', 'c', 'd'])
series_1
```

# 4. Dataframes in Pandas

A 2-dimensional labeled data structure (like an Excel table or SQL table).

Has rows and columns, each column can have a different data type.

```python
# creating a dataframe from dictionary

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
df
```
